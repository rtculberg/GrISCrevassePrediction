# Generate Random Gaussian Fields to Perturb Velocity Maps

## Import Libraries  

This script requires the GStatSim package. Documentation and installation instructions are available here: https://gatorglaciology.github.io/gstatsimbook/intro.html. Scikit-Gstat is also required (https://pypi.org/project/scikit-gstat/). 

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from skgstat import models
import gstools as gs
from gstools import SRF, Exponential, Gaussian
```

### Generate a Test RGF for a Small Area

```python
# exponential test case

import matplotlib.pyplot as plt
#x = y = range(100)
xx, yy = np.meshgrid(np.arange(100), np.arange(100))
model = gs.Exponential(dim=2,
            var= 1,
            len_scale=[15,15],
            angles=0,
            nugget=0
        )

gen = gs.field.generator.RandMeth(model, mode_no=10000, seed=1, sampling='inversion')
field = gen([xx.flatten(),yy.flatten()], add_nugget=False).reshape(xx.shape)

plt.imshow(field)
```

```python
np.shape(field)
```

```python
# save field
np.savetxt(r'Data/random_field_test_exponential.txt', field, delimiter=',')
```

## Generate RGFs for the Whole Greenland Grid

```python
for i in range(10):
    # Generate meshgrid
    xx, yy = np.meshgrid(np.arange(6020), np.arange(10920))   # size of Greenland velocity maps
    
    # Define the model - this uses an exponential covariance model, change to gs.Gaussian for gaussian covariance model
    model = gs.Exponential(dim=2,
                var= 1,
                len_scale=[20,20], # 250 m resolution, 5 km wavelength (so 4*5 = 20)
                angles=0,
                nugget=0
            )

    # Generate random Gaussian field
    gen = gs.field.generator.RandMeth(model, mode_no=10000, seed=i, sampling='inversion')
    field = gen([xx.flatten(),yy.flatten()], add_nugget=False).reshape(xx.shape)
    
    # Save the field to a text file
    np.savetxt(r'Data/field_{}.txt'.format(i), field, delimiter=',')
```
